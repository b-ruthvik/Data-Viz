```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import ipywidgets
#import plotly.express as px
#import plotly.figure_factory as FF
#import plotly.plotly as py
#import plotly.graph_objs as go
from matplotlib.pyplot import figure
```

```python
ufos = pd.read_csv('/Users/ruthvikb/Downloads/ufo-scrubbed-geocoded-time-standardized-00.csv')
```

```python
ufos = ufos.fillna({'City':'unkown', 'Country':'unknown', 'State':'unknown', 'Shape':'unknown' })
```

```python
ufos["Country"].describe()
```

```python
stats = ufos.groupby('Country')["shape"].describe()
```

```python
stats
```

```python
result = ufos.groupby(['Country']).size() 
sns.barplot(x = result.index, y = result.values)
plt.title('Number of UFO sightings by country', weight = 'bold')
plt.xlabel('Country', weight = 'bold')
#plt.yscale("log")
```

```python
fig, ax = plt.subplots(figsize=(10,6))
ax.grid(b = True, color ='grey',
        linestyle ='-.', linewidth = 0.5,
        alpha = 0.85)
ax.bar(result.index, result.values)
ax.set_xlabel('Country', 
               fontweight ='bold')
ax.set_title('Number of UFO sightings by country', fontweight = 'bold')
```

```python
@ipywidgets.interact(style = plt.style.available)
def make_plot(style): 
    with plt.style.context(style):
        fig, ax = plt.subplots(1,1, figsize=(10,6))
        ax.plot(stats["unique"], marker='.', linewidth=1.0, label="unique")
        #ax.plot(stats["min"], marker='.', linewidth=1.0, label="Min")
        ax.set_label('No of unique shapes by Country')
        ax.set_ylabel ("No of unique shapes")
        ax.set_xlabel ("Country")
        #ax.set_yscale("log")
        #ax.legend()
        plt.ylim(0, 40)
    plt.show()
```
